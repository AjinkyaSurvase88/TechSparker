```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
#loading the data file using pandas
custom_dataset = pd.read_csv('Customer Churn.csv')
```

```python
# 5 rows
custom_dataset.head()
```

```python
#no of rows and columns
custom_dataset.shape
```

```python
#geting info about dataset
custom_dataset.info()
```

```python
#checking missing no
custom_dataset.isnull().sum()
```

```python
# Drop rows with any null values
custom_dataset.dropna(inplace=True)
```

```python
# Checking distribution
print(custom_dataset['Status'].value_counts())
```

```python
custom_dataset.head()
```

```python
# Spliting the data into traing and test data
X = custom_dataset.drop(['Churn'],axis = 1)
Y = custom_dataset['Churn']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1,random_state = 2)
```

```python
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

```python
# model evaluation
# prediction of training data
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
# R square Error
error_score = metrics.r2_score(Y_train,training_data_prediction)
print("R square error",error_score)
```

```python
plt.scatter(X_train,training_data_prediction)
plt.ylabel("Actual Role:")
plt.ylabel("Predicted Role:")
plt.title("Actual Role vs Predicted Role")
plt.show()
```
